## Relationship between distance from MRT and monthly rent

From common sense, we believe that with shorter distance to MRT, we are able to fetch higher rental price. In order to compare the rental irrespective of size, we take the rental per sqm instead. 

```python
train_df.head()
```

```python
from preprocessing import distance_to_nearest_place

MRT_DATAFRAME_PATH = "data/auxiliary-data/sg-mrt-existing-stations.csv"
POSITION_ATTRS = ['latitude', 'longitude']
mrt_df = pd.read_csv(MRT_DATAFRAME_PATH).drop_duplicates(POSITION_ATTRS)
train_df[['nearest_mrt_dist', 'nearest_mrt_code']] = \
    train_df.apply(
        lambda row: distance_to_nearest_place(row, existing_mrt_df, 'code'),
        axis=1,
        result_type="expand"
)
```

```python
train_df.head()
```

```python
train_df['rent_per_sq'] = train_df['monthly_rent'] / train_df['floor_area_sqm']
```

```python
train_df.head()
```

```python
flat_types = train_df["flat_type"].unique()
overall_df = train_df[["flat_type", "nearest_mrt_dist", "monthly_rent"]]
for ft in flat_types:
    flat_mrt_rent_df = overall_df[overall_df["flat_type"]==ft]
    flat_mrt_rent_df.plot.scatter(x="nearest_mrt_dist", y="monthly_rent", title=f"{ft}")
```

```python
# train_df[train_df["flat_type"]=="2 room"]["monthly_rent"].idxmax()
# train_df.iloc[23082]
max_list = []
for ft in flat_types:
    max_list.append(train_df.iloc[train_df[train_df["flat_type"]==ft]["monthly_rent"].idxmax()])
max_df = pd.DataFrame(max_list)
max_df.reset_index()[["flat_type", "nearest_mrt_dist", "monthly_rent"]]
```

```python
year2021 = train_df[train_df["rent_approval_date"]<datetime(2022, 1, 1)]
by_flattype_2021 = year2021[["flat_type", "nearest_mrt_dist", "monthly_rent"]].groupby(["flat_type"])
by_flattype_2021 = by_flattype_2021.apply(lambda x: x.sort_values(["nearest_mrt_dist"]))
# by_flattype_2021 = by_flattype_2021[["nearest_mrt_dist","monthly_rent"]].reset_index()[["flat_type", "nearest_mrt_dist", "monthly_rent"]]
by_flattype_2021
```

```python
flat_types = by_flattype_2021["flat_type"].unique()
```
